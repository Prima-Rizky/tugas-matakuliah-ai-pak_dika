```python
import pandas as pd

pd.set_option('display.max_rows', None)     # Menampilkan semua baris
pd.set_option('display.max_columns', None)  # Menampilkan semua kolom

df = pd.read_csv('jumlah fasilitas olahraga samarinda 2022.csv')
df
```

```python
df.info()
```

```python
new_header = df.iloc[2]
df = df[3:].copy()
df.columns = new_header
df.head()
```

```python
df.columns = ['Kelurahan', 'Bulu Tangkis 2022', 'Bola Voli 2022', 'Bola Basket 2022', 'Tenis 2022', 'Sepak Bola 2022', 'Gedung Olahraga 2022', 'Jumlah 2022']
print(df.columns)
```

```python
df.dropna(subset=['Kelurahan'], inplace=True)
df = df[~df['Kelurahan'].str.contains('Jumlah', na=False)]
df = df[df['Jumlah 2022'] != '2022']
df
```

```python
cols_to_convert = ['Bulu Tangkis 2022', 'Bola Voli 2022', 'Bola Basket 2022', 'Tenis 2022', 'Sepak Bola 2022', 'Gedung Olahraga 2022', 'Jumlah 2022']
for col in cols_to_convert:
    df[col] = pd.to_numeric(df[col], errors='coerce')
df.info()
```

```python
df.head()
```

```python
# Binarisasi kolom 'Jumlah 2022'
# Kita akan membuat kolom baru 'Ada Fasilitas Olahraga'
# Nilai 1 jika 'Jumlah 2022' > 0, dan 0 jika 'Jumlah 2022' == 0
df['Ada Fasilitas Olahraga'] = (df['Jumlah 2022'] > 0).astype(int)

# Menampilkan hasil binarisasi
print("DataFrame setelah Binarisasi:")
display(df.head())
```

**Penjelasan Kode Binarisasi:**

Pada bagian ini, kita melakukan proses binarisasi pada kolom 'Jumlah 2022'. Binarisasi adalah teknik transformasi data di mana nilai-nilai numerik diubah menjadi salah satu dari dua nilai diskrit, biasanya 0 dan 1. Dalam konteks ini, kita ingin mengetahui apakah suatu kelurahan memiliki fasilitas olahraga atau tidak, tanpa memperhatikan berapa jumlah pastinya.

Langkah-langkah dalam kode tersebut adalah sebagai berikut:

1.  `df['Ada Fasilitas Olahraga']`: Ini membuat kolom baru di DataFrame kita yang diberi nama 'Ada Fasilitas Olahraga'. Kolom ini akan menyimpan hasil binarisasi.
2.  `(df['Jumlah 2022'] > 0)`: Ini adalah operasi boolean yang membandingkan setiap nilai di kolom 'Jumlah 2022' dengan angka 0. Jika nilai di kolom 'Jumlah 2022' lebih besar dari 0, hasilnya adalah `True`. Jika nilainya 0 atau kurang, hasilnya adalah `False`.
3.  `.astype(int)`: Metode ini digunakan untuk mengubah tipe data dari hasil boolean (`True`/`False`) menjadi integer (`1`/`0`). `True` akan menjadi 1, dan `False` akan menjadi 0. Dengan demikian, kolom 'Ada Fasilitas Olahraga' akan berisi 1 jika kelurahan tersebut memiliki fasilitas olahraga, dan 0 jika tidak.

**Output Binarisasi:**

Output menampilkan beberapa baris pertama dari DataFrame setelah kolom 'Ada Fasilitas Olahraga' ditambahkan. Anda dapat melihat bahwa kolom baru ini berisi nilai 0 atau 1, yang menunjukkan keberadaan fasilitas olahraga di setiap kelurahan.

```python
from sklearn.preprocessing import MinMaxScaler

# Pilih 3 kolom numerik untuk normalisasi
# Sebagai contoh, kita pilih 'Bulu Tangkis 2022', 'Bola Voli 2022', dan 'Sepak Bola 2022'
cols_to_normalize = ['Bulu Tangkis 2022', 'Bola Voli 2022', 'Sepak Bola 2022']

# Inisialisasi MinMaxScaler untuk rentang [-1, 1]
scaler = MinMaxScaler(feature_range=(-1, 1))

# Lakukan normalisasi pada 3 kolom yang dipilih
df[cols_to_normalize] = scaler.fit_transform(df[cols_to_normalize])

# Menampilkan hasil normalisasi
print("DataFrame setelah Normalisasi (rentang -1 hingga 1) pada 3 kolom terpilih:")
display(df.head())
```

**Penjelasan Kode Normalisasi:**

Selanjutnya, kita melakukan normalisasi data menggunakan metode Min-Max Scaling dengan rentang nilai antara -1 dan 1, namun kali ini hanya pada 3 kolom numerik yang kita pilih. Normalisasi penting untuk menyamakan skala fitur-fitur yang berbeda sehingga tidak ada satu fitur pun yang mendominasi dalam analisis atau model berbasis jarak.

Langkah-langkah dalam kode tersebut adalah sebagai berikut:

1.  `from sklearn.preprocessing import MinMaxScaler`: Mengimpor kelas `MinMaxScaler` dari modul `sklearn.preprocessing`.
2.  `cols_to_normalize`: Mendefinisikan daftar yang hanya berisi 3 nama kolom numerik yang akan dinormalisasi sesuai permintaan. Dalam contoh ini, kita memilih 'Bulu Tangkis 2022', 'Bola Voli 2022', dan 'Sepak Bola 2022'. Anda bisa mengganti nama kolom ini sesuai kebutuhan analisis Anda.
3.  `scaler = MinMaxScaler(feature_range=(-1, 1))`: Menginisialisasi objek `MinMaxScaler` dengan `feature_range` diatur ke `(-1, 1)`. Ini memastikan bahwa setelah normalisasi, semua nilai dalam kolom yang dipilih akan berada dalam rentang tertutup dari -1 hingga 1.
4.  `df[cols_to_normalize] = scaler.fit_transform(df[cols_to_normalize])`: Ini adalah langkah di mana normalisasi diterapkan.
    *   `scaler.fit(df[cols_to_normalize])`: Objek `scaler` mempelajari skala (nilai minimum dan maksimum) dari data di 3 kolom yang dipilih.
    *   `scaler.transform(df[cols_to_normalize])`: Kemudian, data di 3 kolom tersebut diubah (diskalakan) menggunakan rumus Min-Max Scaling agar sesuai dengan rentang [-1, 1]. Rumusnya adalah:
