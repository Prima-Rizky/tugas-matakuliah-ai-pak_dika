```python
import pandas as pd

pd.set_option('display.max_rows', None)     # Menampilkan semua baris
pd.set_option('display.max_columns', None)  # Menampilkan semua kolom

df = pd.read_csv('jumlah fasilitas olahraga samarinda 2022.csv')
df
```

```python
df.info()
```

```python
new_header = df.iloc[2]
df = df[3:].copy()
df.columns = new_header
df.head()
```

```python
df.columns = ['Kelurahan', 'Bulu Tangkis 2022', 'Bola Voli 2022', 'Bola Basket 2022', 'Tenis 2022', 'Sepak Bola 2022', 'Gedung Olahraga 2022', 'Jumlah 2022']
print(df.columns)
```

```python
df.dropna(subset=['Kelurahan'], inplace=True)
df = df[~df['Kelurahan'].str.contains('Jumlah', na=False)]
df = df[df['Jumlah 2022'] != '2022']
df
```

```python
cols_to_convert = ['Bulu Tangkis 2022', 'Bola Voli 2022', 'Bola Basket 2022', 'Tenis 2022', 'Sepak Bola 2022', 'Gedung Olahraga 2022', 'Jumlah 2022']
for col in cols_to_convert:
    df[col] = pd.to_numeric(df[col], errors='coerce')
df.info()
```

```python
df.head()
```

```python
# Binarisasi kolom 'Jumlah 2022'
# Kita akan membuat kolom baru 'Ada Fasilitas Olahraga'
# Nilai 1 jika 'Jumlah 2022' > 0, dan 0 jika 'Jumlah 2022' == 0
df['Ada Fasilitas Olahraga'] = (df['Jumlah 2022'] > 0).astype(int)

# Menampilkan hasil binarisasi
print("DataFrame setelah Binarisasi:")
display(df.head())
```

**Penjelasan Kode Binarisasi:**

Pada bagian ini, kita melakukan proses binarisasi pada kolom 'Jumlah 2022'. Binarisasi adalah teknik transformasi data di mana nilai-nilai numerik diubah menjadi salah satu dari dua nilai diskrit, biasanya 0 dan 1. Dalam konteks ini, kita ingin mengetahui apakah suatu kelurahan memiliki fasilitas olahraga atau tidak, tanpa memperhatikan berapa jumlah pastinya.

Langkah-langkah dalam kode tersebut adalah sebagai berikut:

1.  `df['Ada Fasilitas Olahraga']`: Ini membuat kolom baru di DataFrame kita yang diberi nama 'Ada Fasilitas Olahraga'. Kolom ini akan menyimpan hasil binarisasi.
2.  `(df['Jumlah 2022'] > 0)`: Ini adalah operasi boolean yang membandingkan setiap nilai di kolom 'Jumlah 2022' dengan angka 0. Jika nilai di kolom 'Jumlah 2022' lebih besar dari 0, hasilnya adalah `True`. Jika nilainya 0 atau kurang, hasilnya adalah `False`.
3.  `.astype(int)`: Metode ini digunakan untuk mengubah tipe data dari hasil boolean (`True`/`False`) menjadi integer (`1`/`0`). `True` akan menjadi 1, dan `False` akan menjadi 0. Dengan demikian, kolom 'Ada Fasilitas Olahraga' akan berisi 1 jika kelurahan tersebut memiliki fasilitas olahraga, dan 0 jika tidak.

**Output Binarisasi:**

Output menampilkan beberapa baris pertama dari DataFrame setelah kolom 'Ada Fasilitas Olahraga' ditambahkan. Anda dapat melihat bahwa kolom baru ini berisi nilai 0 atau 1, yang menunjukkan keberadaan fasilitas olahraga di setiap kelurahan.

```python
from sklearn.preprocessing import MinMaxScaler

# Pilih 3 kolom numerik untuk normalisasi
# Sebagai contoh, kita pilih 'Bulu Tangkis 2022', 'Bola Voli 2022', dan 'Sepak Bola 2022'
cols_to_normalize = ['Bulu Tangkis 2022', 'Bola Voli 2022', 'Sepak Bola 2022']

# Inisialisasi MinMaxScaler untuk rentang [-1, 1]
scaler = MinMaxScaler(feature_range=(-1, 1))

# Lakukan normalisasi pada 3 kolom yang dipilih
df[cols_to_normalize] = scaler.fit_transform(df[cols_to_normalize])

# Menampilkan hasil normalisasi
print("DataFrame setelah Normalisasi (rentang -1 hingga 1) pada 3 kolom terpilih:")
display(df.head())
```

**Penjelasan Kode Normalisasi:**

Selanjutnya, kita melakukan normalisasi data menggunakan metode Min-Max Scaling dengan rentang nilai antara -1 dan 1, namun kali ini hanya pada 3 kolom numerik yang kita pilih. Normalisasi penting untuk menyamakan skala fitur-fitur yang berbeda sehingga tidak ada satu fitur pun yang mendominasi dalam analisis atau model berbasis jarak.

Langkah-langkah dalam kode tersebut adalah sebagai berikut:

1.  `from sklearn.preprocessing import MinMaxScaler`: Mengimpor kelas `MinMaxScaler` dari modul `sklearn.preprocessing`.
2.  `cols_to_normalize`: Mendefinisikan daftar yang hanya berisi 3 nama kolom numerik yang akan dinormalisasi sesuai permintaan. Dalam contoh ini, kita memilih 'Bulu Tangkis 2022', 'Bola Voli 2022', dan 'Sepak Bola 2022'. Anda bisa mengganti nama kolom ini sesuai kebutuhan analisis Anda.
3.  `scaler = MinMaxScaler(feature_range=(-1, 1))`: Menginisialisasi objek `MinMaxScaler` dengan `feature_range` diatur ke `(-1, 1)`. Ini memastikan bahwa setelah normalisasi, semua nilai dalam kolom yang dipilih akan berada dalam rentang tertutup dari -1 hingga 1.
4.  `df[cols_to_normalize] = scaler.fit_transform(df[cols_to_normalize])`: Ini adalah langkah di mana normalisasi diterapkan.
    *   `scaler.fit(df[cols_to_normalize])`: Objek `scaler` mempelajari skala (nilai minimum dan maksimum) dari data di 3 kolom yang dipilih.
    *   `scaler.transform(df[cols_to_normalize])`: Kemudian, data di 3 kolom tersebut diubah (diskalakan) menggunakan rumus Min-Max Scaling agar sesuai dengan rentang [-1, 1]. Rumusnya adalah:

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Pilih kolom yang sudah dinormalisasi dan kolom 'Kelurahan'
# Asumsi df dan cols_to_normalize sudah tersedia dari langkah sebelumnya
df_normalized_viz = df[['Kelurahan'] + cols_to_normalize]

# Membuat bar plot untuk 3 kolom yang dinormalisasi
df_melted = df_normalized_viz.melt(id_vars='Kelurahan', var_name='Jenis Fasilitas', value_name='Nilai Normalisasi')

plt.figure(figsize=(15, 8))
sns.barplot(x='Kelurahan', y='Nilai Normalisasi', hue='Jenis Fasilitas', data=df_melted)
plt.xticks(rotation=90)
plt.title('Nilai Normalisasi Fasilitas Olahraga per Kelurahan (Rentang -1 hingga 1)')
plt.ylabel('Nilai Normalisasi')
plt.xlabel('Kelurahan')
plt.tight_layout()
plt.show()
```

**Penjelasan Visualisasi:**

Visualisasi di atas menunjukkan nilai normalisasi dari tiga jenis fasilitas olahraga (Bulu Tangkis, Bola Voli, dan Sepak Bola) untuk setiap kelurahan. Normalisasi dilakukan menggunakan Min-Max Scaling dengan rentang nilai antara -1 dan 1.

* **Sumbu X:** Menunjukkan nama kelurahan.
* **Sumbu Y:** Menunjukkan nilai normalisasi dari jumlah fasilitas olahraga. Nilai 1 menunjukkan jumlah fasilitas tertinggi di antara semua kelurahan untuk jenis fasilitas tersebut, nilai -1 menunjukkan jumlah terendah (atau 0 jika nilai minimum asli adalah 0), dan 0 menunjukkan nilai di tengah rentang.
* **Warna Batang:** Setiap warna batang mewakili jenis fasilitas olahraga yang berbeda (Bulu Tangkis, Bola Voli, Sepak Bola).

Dari visualisasi ini, kita dapat membandingkan secara relatif jumlah fasilitas olahraga di setiap kelurahan untuk ketiga jenis tersebut setelah diskalakan ke rentang yang sama. Batang yang lebih tinggi menunjukkan kelurahan dengan jumlah fasilitas yang relatif lebih banyak dibandingkan kelurahan lain untuk jenis fasilitas tersebut, dan sebaliknya.

Visualisasi ini membantu kita melihat pola distribusi fasilitas olahraga di Samarinda pada tahun 2022 dari perspektif nilai yang dinormalisasi, yang berguna untuk analisis lebih lanjut atau pemodelan yang sensitif terhadap skala data.

```python
# Step 1: Create 'Kecamatan' column
df['Kecamatan'] = df['Kelurahan'].apply(lambda x: x.split(' [')[0] if '[' in x else None)

# Fill the None values (sub-kelurahan) with the previous non-None value (kecamatan)
df['Kecamatan'] = df['Kecamatan'].fillna(method='ffill')

# Remove leading and trailing whitespace and the indentation character
df['Kecamatan'] = df['Kecamatan'].str.strip().str.replace('  ', '', regex=False)


# Step 2 & 3: Group by 'Kecamatan' and sum 'Jumlah 2022'
kecamatan_facilities = df.groupby('Kecamatan')['Jumlah 2022'].sum().reset_index()

# Step 4: Create a bar plot
plt.figure(figsize=(12, 6))
sns.barplot(x='Kecamatan', y='Jumlah 2022', data=kecamatan_facilities)
plt.xticks(rotation=45, ha='right')
plt.title('Total Jumlah Fasilitas Olahraga per Kecamatan di Samarinda (2022)')
plt.ylabel('Jumlah Total Fasilitas')
plt.xlabel('Kecamatan')
plt.tight_layout()
plt.show()
```

```python
# Select the columns representing the different sports facility types
facility_cols = ['Bulu Tangkis 2022', 'Bola Voli 2022', 'Bola Basket 2022', 'Tenis 2022', 'Sepak Bola 2022', 'Gedung Olahraga 2022']

# Calculate the sum of facilities for each selected type
total_facilities_per_type = df[facility_cols].sum()

# Create a pandas Series from the calculated sums
facilities_distribution = pd.Series(total_facilities_per_type)

# Create a bar plot of the Series to visualize the distribution
plt.figure(figsize=(10, 6))
facilities_distribution.plot(kind='bar')

# Add title and labels
plt.title('Distribusi Jumlah Fasilitas Olahraga berdasarkan Jenis di Samarinda (2022)')
plt.ylabel('Jumlah Total Fasilitas')
plt.xlabel('Jenis Fasilitas')

# Rotate x-axis labels for readability
plt.xticks(rotation=45, ha='right')

# Ensure the layout is tight
plt.tight_layout()

# Display the plot
plt.show()
```

```python
# Select the original columns representing the different sports facility types before normalization
original_facility_cols = ['Bulu Tangkis 2022', 'Bola Voli 2022', 'Bola Basket 2022', 'Tenis 2022', 'Sepak Bola 2022', 'Gedung Olahraga 2022']

# Select the columns representing the different sports facility types that are numeric
numeric_facility_cols = ['Bulu Tangkis 2022', 'Bola Voli 2022', 'Bola Basket 2022', 'Tenis 2022', 'Sepak Bola 2022', 'Gedung Olahraga 2022']

# Calculate the sum of facilities for each selected type
total_facilities_per_type_numeric = df[numeric_facility_cols].sum()

# Create a pandas Series from the calculated sums
facilities_distribution_numeric = pd.Series(total_facilities_per_type_numeric)

# Create a bar plot of the Series to visualize the distribution
plt.figure(figsize=(10, 6))
facilities_distribution_numeric.plot(kind='bar')

# Add title and labels
plt.title('Distribusi Jumlah Fasilitas Olahraga berdasarkan Jenis di Samarinda (2022)')
plt.ylabel('Jumlah Total Fasilitas')
plt.xlabel('Jenis Fasilitas')

# Rotate x-axis labels for readability
plt.xticks(rotation=45, ha='right')

# Ensure the layout is tight
plt.tight_layout()

# Display the plot
plt.show()
```
